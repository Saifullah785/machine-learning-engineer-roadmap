# **‚úÖ What Are Mixed Variables in the Same Column?**

Mixed variables in the same column mean:

A single column contains different types of values ‚Äì e.g., numbers, categories, or messy text.

**üîç Example in real life:**

Imagine a column called rating:

Some rows: "Excellent", "Good"

Other rows: 5, 4

Some: "N/A" or "-"

This is a dirty column and needs cleaning + separation.

**üìå Goal of Feature Engineering:**

Detect inconsistent formats.

Convert everything into a clean, structured form that ML models understand.

Often we split, clean, or transform such columns.



```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with ')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python

# extract categorical part


df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python

df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part

df['cabin_cat'] = df['Cabin'].str[0]      # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot.bar()
```

```python
# extract the last bit of ticket as number

df['ticket_num'] = df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head()
```
