# **3.Opening a csv file from an URL**

```python
import requests
from io import StringIO
Url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day15%20-%20working%20with%20csv%20files/aug_train.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'}
req = requests.get(Url,headers=headers)
data = StringIO(req.text)
df=pd.read_csv(data)
df.head()
```

# **14. Convertors**

```python
def rename(name):
  if name == "Royal Challengers Bangalore":
    return "RCB"
  else:
    return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
df = pd.read_csv('/content/IPL Matches 2008-2020.csv',converters={'team1':rename})
```

```python
df.head()
```
