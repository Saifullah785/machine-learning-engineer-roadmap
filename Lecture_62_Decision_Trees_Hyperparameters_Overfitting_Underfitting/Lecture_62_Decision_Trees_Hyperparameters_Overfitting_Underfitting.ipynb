```python
import numpy as np # Importing the numpy library for numerical operations

import pandas as pd # Importing the pandas library for data manipulation

import matplotlib.pyplot as plt # Importing the matplotlib library for plotting

from sklearn.tree import DecisionTreeClassifier # Importing DecisionTreeClassifier from scikit-learn for building the decision tree model
```

```python
def analyzer(max_depth): # Defining a function called analyzer that takes max_depth as an argument

  data = pd.read_csv('/content/Social_Network_Ads.csv') # Reading the data from the csv file into a pandas DataFrame

  x = data.iloc[:,2:4].values # Selecting the features (columns 2 and 3) and converting them to a numpy array

  y = data.iloc[:,-1].values # Selecting the target variable (last column) and converting it to a numpy array

  clf = DecisionTreeClassifier(max_depth=max_depth) # Creating a Decision Tree Classifier model with the specified max_depth

  clf.fit(x,y) # Training the model on the features (x) and target variable (y)

  a = np.arange(start=x[:,0].min()-1, stop=x[:,0].max()+1, step=0.1) # Creating a range of values for the first feature

  b = np.arange(start=x[:,1].min()-1, stop=x[:,1].max()+1, step=100) # Creating a range of values for the second feature


  XX, YY = np.meshgrid(a,b) # Creating a meshgrid from the ranges of the two features

  input_array = np.array([XX.ravel(), YY.ravel()]).T # Creating an input array for prediction using the meshgrid values

  labels = clf.predict(input_array) # Predicting the class labels for the input array using the trained model

  plt.contourf(XX, YY, labels.reshape(XX.shape), alpha=0.5) # Plotting the decision boundary using contourf

  plt.scatter(x[:,0], x[:,1], c=y) # Plotting the original data points with different colors based on their class labels

  plt.show() # Displaying the plot
```

```python
analyzer(max_depth=1) # Calling the analyzer function with max_depth set to 1
```

```python
analyzer(max_depth=2) # Calling the analyzer function with max_depth set to 2
```

```python
analyzer(max_depth=3) # Calling the analyzer function with max_depth set to 3
```

```python
analyzer(max_depth=5) # Calling the analyzer function with max_depth set to 5
```

```python
analyzer(max_depth=None) # Calling the analyzer function with max_depth set to None (allowing the tree to grow fully)
```
