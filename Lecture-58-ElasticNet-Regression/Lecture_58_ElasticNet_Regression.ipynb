# **Purpose:**

The notebook demonstrates how to load a standard dataset (Diabetes), prepare it for machine learning, and train and evaluate different linear regression models on the data.

# **Data Loading and Splitting:**

It starts by loading the diabetes dataset, separating the features (input variables) from the target variable (the outcome to be predicted). It then splits this data into training and testing sets to allow for model evaluation on unseen data.

# **Model Training:**

The notebook trains four different types of linear regression models: standard Linear Regression, Ridge Regression, Lasso Regression, and Elastic Net Regression. These models are trained on the training data to learn the relationships between the features and the target variable.

# **Model Evaluation:**

After training, each model is used to make predictions on the testing data. The performance of each model is then evaluated using the R-squared score, which measures how well the model's predictions fit the actual target values.

# **Data Exploration (Optional):**

The notebook includes sections to explore the loaded data using pandas DataFrames and visualizations (histograms and scatter plots) to understand the structure and distribution of the data and the relationship between features and the target variable.

```python
# Imports necessary libraries for data manipulation, machine learning, and visualization.
# Imports specific modules for loading the dataset, different linear regression models,
# splitting data, and evaluating model performance.

from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
# Loads the diabetes dataset, separating features (X) and the target variable (y).

X,y = load_diabetes(return_X_y=True)
```

```python
X
```

```python
# Splits the dataset into training and testing sets (80% train, 20% test) for model evaluation.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
# Initializes and trains a Linear Regression model on the training data.
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Makes predictions on the test data using the trained Linear Regression model
# and calculates the R-squared score.

y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```

```python
# Initializes and trains a Ridge Regression model with a specified alpha value on the training data.

reg = Ridge(alpha=0.1)
reg.fit(X_train, y_train)
```

```python
# Makes predictions on the test data using the trained Ridge Regression model

y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```

```python
# Initializes and trains a Lasso Regression model with a specified alpha value on the training data.

reg = Lasso(alpha=0.01)
reg.fit(X_train, y_train)
```

```python
# Makes predictions on the test data using the trained Lasso Regression model
# and calculates the R-squared score.

y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```

```python
# Initializes and trains an Elastic Net Regression model with specified alpha and l1_ratio values on the training data.
# Elastic Net

reg = ElasticNet(alpha=0.005, l1_ratio=0.9)
reg.fit(X_train, y_train)
```

```python
# prediction
y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```
