```python
# Import necessary libraries for data manipulation, machine learning models, and plotting.
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Generate a synthetic classification dataset with 5 features, all of which are informative.
X, y = make_classification(n_features=5, n_informative=5, n_redundant=0, n_clusters_per_class=1)
```

```python
# Create a pandas DataFrame from the generated data.
# Name the feature columns 'col1' through 'col5' and the target column 'target'.
df = pd.DataFrame(X, columns=['col1','col2', 'col3','col4', 'col5'])
df['target'] = y
# Print the shape of the DataFrame (number of rows and columns).
print(df.shape)
# Display the first few rows of the DataFrame to inspect the data.
df.head()
```

```python
# Initialize a BaggingClassifier model.
# max_features=2 means that each base estimator (decision tree) will consider only 2 random features for each split.
bag = BaggingClassifier(max_features=2)
```

```python
# Train the BaggingClassifier model.
# We use all columns except the last one as features (df.iloc[:,:5]) and the last column as the target (df.iloc[:, -1]).
bag.fit(df.iloc[:,:5],df.iloc[:, -1])
```

```python
# Plot the first decision tree estimator from the BaggingClassifier model.
plt.figure(figsize=(12,12)) # Set the figure size for better visualization.
plot_tree(bag.estimators_[0]) # Plot the tree structure.
plt.show() # Display the plot.
```

```python
# Initialize a RandomForestClassifier model.
# max_features=2 means that each base estimator (decision tree) will consider only 2 random features for each split.
rf = RandomForestClassifier(max_features=2)
```

```python
# Train the RandomForestClassifier model.
# We use all columns except the last one as features (df.iloc[:,:5]) and the last column as the target (df.iloc[:, -1]).
rf.fit(df.iloc[:,:5],df.iloc[:, -1])
```

```python
# Plot the fifth decision tree estimator from the RandomForestClassifier model.
plt.figure(figsize=(12,12)) # Set the figure size for better visualization.
plot_tree(rf.estimators_[4]) # Plot the tree structure.
plt.show() # Display the plot.
```
