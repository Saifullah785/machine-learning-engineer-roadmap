# ✅ What is a Tensor?


---



Tensors are just containers for numbers — like boxes holding data.


They are multidimensional arrays (think of tables, cubes, etc.) that store data in different shapes and dimensions.


✨ A tensor is to ML what ingredients are to cooking.

# ✍️ Code Sample: Creating Tensors

Using NumPy and PyTorch (optional knowledge):

```python
import numpy as np

import torch


# 1D tensor (vector)

a = np.array([1, 2, 3])  # NumPy
print(a)
b = torch.tensor([1, 2, 3])  # PyTorch
print(b)

# 2D tensor (matrix)

matrix = torch.tensor([[1, 2], [3, 4]])
print(matrix)

# 3D tensor (e.g., RGB image)

image_tensor = torch.rand(3, 28, 28)  # 3 channels, 28x28 pixels
print(image_tensor)
```
