```python
# Loop through multiple pages of the API to fetch top-rated movie data
for i in range(1, 429):
    # Send a GET request to the API for each page, including the API key and page number
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=c585f954aa345da200ef56fe75e5c1f4&language=en-US&page={}'.format(i))
    # Create a temporary DataFrame from the 'results' key of the JSON response, selecting specific columns for movies.
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    # Concatenate the temporary DataFrame to the main DataFrame 'df', ignoring the index
    df = pd.concat([df, temp_df], ignore_index=True)
```
